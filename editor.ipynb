```python
import numpy as np
import pandas as pd
```

# Comeng Data

```python
comeng = pd.read_json('Comeng.json')
```

```python
comeng = comeng[comeng["Status:"] != "Scrapped"]
```

```python
comeng = comeng.drop(columns=['In Service:', 'Status:'])
```

```python
comeng = comeng.rename(columns={'Set:': "id", 'Interior:':"notes", 'Notes:':"other", "Livery:":"livery"})
```

```python
comeng.loc[comeng['other'] == 'EED Trial - Modified Horn Installed (Single Tone)', 'notes'] += ', EED Trial'
```

```python
comeng = comeng.drop(columns=['other'])
```

```python
comeng.set_index('id', inplace=True)
```

```python
comeng['train'] = 'Comeng'
```

```python
comeng
```

# Seimens Data

```python
seimens = pd.read_json('Seimens.json')
seimens.head()
```

```python
seimens = seimens[seimens["Status:"] != "Scrapped"]
```

```python
seimens = seimens.drop(columns=['Unit:', 'In Service:', 'Status:'])
```

```python
seimens = seimens.rename(columns={'Set:':"id", 'Livery:':"livery", 'Notes:':"notes"})
```

```python
seimens['train'] = 'Seimens'
```

```python
seimens.set_index('id', inplace=True)
```

```python
seimens
```

# X'Trapolis

```python
xtrap = pd.read_json('XTrapolis.json')
```

```python
xtrap = xtrap[xtrap["Status:"] != "Scrapped"]
```

```python
xtrap = xtrap.drop(columns=['Unit:', 'In Service:', 'Status:'])
```

```python
xtrap = xtrap.rename(columns={'Set:':"id", "Notes:": "notes", 'Livery:':"livery" })
```

```python
xtrap['train'] = "X'Trapolis"
```

```python
xtrap.set_index('id', inplace=True)
```

```python
xtrap
```

# HCMT

```python
hcmt = pd.read_json('HCMT.json')
```

```python
hcmt = hcmt[hcmt["Status:"] != "Scrapped"]
```

```python
hcmt = hcmt.drop(columns=['Unit:', 'In Service:', 'Status:'])
```

```python
hcmt = hcmt.rename(columns={'Set:':"id", "Notes:": "notes", 'Livery:':"livery" })
```

```python
hcmt['train'] = "HCMT"
```

```python
hcmt['id'] = hcmt['id'].str.replace(' ', '', regex=False)
```

```python
hcmt.set_index('id', inplace=True)
```

```python
hcmt
```

# Combining Trains

```python
data = pd.concat([comeng, seimens], ignore_index=False)
data = pd.concat([data, xtrap], ignore_index=False)
data = pd.concat([data, hcmt], ignore_index=False)
```

```python
data
```

```python
json_result = data.reset_index().to_json(orient='records')
```

```python
with open('MelbourneTrains.json', 'w') as file:
    file.write(json_result)
```
